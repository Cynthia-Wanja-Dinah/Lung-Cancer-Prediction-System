```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the dataset
df = pd.read_csv('survey lung cancer.csv')
df
```

```python
df['GENDER'].value_counts()
```

**From the dataset there were more men than female**

```python
#generating counts
df['LUNG_CANCER'].value_counts()
```

```python
#COUNT PLOT FOR LUNG_CANCER
plt.figure(figsize=(8, 6))
sns.countplot(x='LUNG_CANCER', data=df)
plt.title('Count of Lung Cancer')
plt.show()
```

*From the Bar plot we can see that we had most results turned to be positive*

```python
#Distribution Plots for AGE
plt.figure(figsize=(10, 6))
sns.histplot(df['AGE'], kde=True)
plt.title('Age Distribution')
plt.show()

# Distribution of AGE by LUNG_CANCER
plt.figure(figsize=(10, 6))
sns.boxplot(x='LUNG_CANCER', y='AGE', data=df)
plt.title('Age Distribution by Lung Cancer')
plt.show()
```

```python
df.drop(columns ='GENDER', inplace=True)
```

***we had to drop gender because it had no effect to our model***

```python
#changing TReatment Outcome to be a ) and 1
df['LUNG_CANCER'] = df['LUNG_CANCER'].map({'YES': 1, 'NO': 0})
```

```python
plt.figure(figsize=(30,20))
cor = df.corr()
sns.heatmap(cor,vmax=1,square = True,annot=True, cmap=plt.cm.Blues)
plt.title('Correlation between different attributes')
plt.show()
```

**We can see from the heatmap that  there was no correlation among the variable  in the data**

**From the visualizations, it is clear that in the given dataset, the features GENDER, AGE, SMOKING and SHORTNESS OF BREATH don't have that much relationship with LUNG CANCER. So let's drop those features to make this dataset more clean.**
